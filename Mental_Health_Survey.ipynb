**PROJECT SUMMARY**
Mental Health in Tech Survey EDA

This project analyzes a comprehensive mental health survey dataset, exploring employees’ experiences, perceptions, and behaviors related to mental health in the workplace. The dataset includes demographic variables (such as age, gender, country, and employment status), workplace characteristics (company size, industry type, remote work), and detailed questions on mental health awareness, stigma, treatment-seeking, and employer support.

Key insights derived from the data include:

•	**Geographical differences**: Among the top 10 countries represented, New Zealand exhibits the highest proportion of employees seeking mental health treatment, suggesting greater openness, accessibility, or reduced stigma toward mental health care.

•	**Stigma and fear**: A substantial share of respondents believe there would be negative consequences to disclosing mental health issues at work, highlighting ongoing stigma in many workplaces.

•	**Family history and treatment**: Individuals with a family history of mental illness are more likely to seek treatment themselves, indicating that personal awareness and experience may motivate help-seeking behavior.

•	**Industry insights**: Contrary to common assumptions, non-tech companies report offering more mental health benefits than tech companies, challenging the stereotype of tech firms as leaders in employee mental health support.

•	**Benefits and treatment**: Employees with access to mental health benefits from their employers are more likely to seek treatment, underscoring the importance of formal support systems.

•	**Anonymity and treatment**: The dataset suggests that perceived anonymity in seeking treatment does not significantly influence whether employees pursue mental health care, pointing toward other factors playing larger roles.

•	**Self-employment challenges**: Self-employed workers report higher levels of mental health interference with work, highlighting a potentially underserved group in need of tailored support.

•	**Company size and awareness**: Employees in smaller companies tend to be less aware of available mental health care options, suggesting a need for better communication and resources in smaller organizations.

•	**Mental vs. physical health prioritization**: Many employees are uncertain about whether their employer takes mental health as seriously as physical health, revealing a gap in awareness that organizations should address to foster stronger mental health support.

Overall, these findings emphasize the complex interplay of demographic, workplace, and cultural factors in shaping mental health experiences at work. They highlight critical areas for improvement, such as reducing stigma, expanding mental health benefits, enhancing communication—especially in smaller companies—and better supporting self-employed individuals.

**DATA WRANGLING**

```python
#Dealing with null values
#Relacing null values in work_interfere column
df['work_interfere'].fillna("Not Applicable", inplace=True)
```

```python
#Work_interfere column after relacing nulls.
df['work_interfere'].isnull().sum()
```

```python
#clean no_employyes column
df['no_employees'] = df['no_employees'].astype(str).str.strip()
import re

def is_valid_size(value):
    # List of known valid ranges
    valid_sizes = [
        '1-5', '6-25', '26-100', '100-500',
        '500-1000', '1000-5000', '5000-10000',
        'More than 10000', 'More than 1000'
    ]
    # Clean the value first
    value = str(value).strip()
    # Filter out date-like entries (e.g. "25-Jun", "5-Jan")
    if re.match(r'^\d{1,2}-[A-Za-z]{3}$', value):
        return False
    return value in valid_sizes

# Apply filter
df = df[df['no_employees'].apply(is_valid_size)]
def size_group(size):
    if size in ['26-100']:
        return 'Small'
    elif size in ['100-500']:
        return 'Medium'
    elif size in ['500-1000', '1000-5000', '5000-10000', 'More than 10000', 'More than 1000']:
        return 'Large'
    else:
        return 'Unknown'

df['company_size_group'] = df['no_employees'].apply(size_group)
print(df['no_employees'].value_counts())
```

1)How does the frequency of mental health illness and attitudes towards mental health vary by geographic locations?

```python
top_countries = df['Country'].value_counts().head(10).index
df_top = df[df['Country'].isin(top_countries)]

treatment_rate = df_top[df_top['treatment'].notna()].groupby('Country')['treatment'].value_counts(normalize=True).unstack().fillna(0)

treatment_rate['Yes'].sort_values().plot(kind='barh', color='skyblue')
plt.xlabel('Proportion of People Who Sought Treatment')
plt.title('Mental Health Treatment by Country')
plt.tight_layout()
plt.show()
```

By Attitude

```python
attitudes = (
    df_top.groupby('Country')['mental_health_consequence']
    .value_counts(normalize=True)
    .unstack()
    .fillna(0)
    .sort_values(by='Yes', ascending=False)
)

# Step 3: Plot
plt.figure(figsize=(7, 4.5))
sns.barplot(x=attitudes['Yes'], y=attitudes.index, palette='Reds')

plt.xlabel('% Who Believe There Are Negative Consequences')
plt.ylabel('Country')
plt.title('⚠️ Attitudes Toward Mental Health Disclosure at Work (by Country)')
plt.xlim(0, 1)
plt.tight_layout()
plt.show()
#This shows the proportion of respondents in each country who believe there would be negative consequences to disclosing a mental health issue at work —
#a good measure of stigma or fear.
```

2)What are the strongest predictors of mental health illness or certain attitudes towards mental health in the workplace?

```python
# Clean 'treatment' column: ensure consistent format and filter only Yes/No
df['treatment'] = df['treatment'].astype(str).str.strip().str.capitalize()
df = df[df['treatment'].isin(['Yes', 'No'])]

# Map 'treatment' to numeric for averaging
df['treatment_num'] = df['treatment'].map({'Yes': 1, 'No': 0})

# Group by family_history and calculate mean treatment rate
treatment_rate = df.groupby('family_history')['treatment_num'].mean()

# Plot bar chart
treatment_rate.plot(kind='bar', color=['skyblue', 'salmon'])
plt.title('Proportion Seeking Mental Health Treatment by Family History')
plt.xlabel('Family History of Mental Illness')
plt.ylabel('Proportion Seeking Treatment')
plt.xticks(ticks=[0,1], labels=['No', 'Yes'], rotation=0)
plt.ylim(0,1)
plt.show()
```

3)Do tech companies provide better mental health support than non-tech companies?

```python
# Filter data to only Yes/No responses for 'benefits'
df_simple = df[df['benefits'].isin(['Yes', 'No'])]

# Calculate % who said 'Yes' by tech company
support_rate = df_simple.groupby('tech_company')['benefits'].apply(lambda x: (x == 'Yes').mean())

# Plot
support_rate.plot(kind='bar', color=['lightgray', 'skyblue'], figsize=(5,4))

plt.xticks([0,1], ['Non-Tech', 'Tech'], rotation=0)
plt.ylabel('Proportion with Mental Health Benefits')
plt.title('Mental Health Benefits by Company Type')
plt.ylim(0,1)
plt.tight_layout()
plt.show()
```

4)Does the presence of mental health benefits influence whether employees seek treatment?
[Compare the % of employees who sought treatment grouped by whether their employer provides mental health benefits.]

```python
# Filter to rows with valid 'benefits' and 'treatment' responses
df_valid = df[df['benefits'].isin(['Yes', 'No']) & df['treatment'].isin(['Yes', 'No'])]

# Calculate % who sought treatment grouped by 'benefits'
treatment_rate = df_valid.groupby('benefits')['treatment'].apply(lambda x: (x == 'Yes').mean())

# Plot
treatment_rate.plot(kind='bar', color=['salmon', 'seagreen'], figsize=(6,4))

plt.xticks(rotation=0)
plt.xlabel('Employer Provides Mental Health Benefits')
plt.ylabel('Proportion Who Sought Treatment')
plt.title('Impact of Mental Health Benefits on Seeking Treatment')
plt.ylim(0,1)
plt.tight_layout()
plt.show()
```

5)How does the perception of anonymity affect the likelihood of seeking help?[Compare the percentage of employees who sought treatment based on their response to anonymity protection (Yes/No/Don’t know).]

```python
# Filter for valid responses
df_filtered = df[
    df['anonymity'].isin(['Yes', 'No', "Don't know"]) &
    df['treatment'].isin(['Yes', 'No'])
]

# Calculate proportion who sought treatment by anonymity response
treatment_by_anon = df_filtered.groupby('anonymity')['treatment'].apply(lambda x: (x == 'Yes').mean())

# Plot
treatment_by_anon.plot(kind='bar', color=['seagreen', 'salmon', 'gray'], figsize=(6,4))

plt.title('Effect of Perceived Anonymity on Seeking Treatment')
plt.xlabel('Anonymity Protected When Using Mental Health Resources')
plt.ylabel('Proportion Who Sought Treatment')
plt.ylim(0,1)
plt.xticks(rotation=0)
plt.tight_layout()
plt.show()
```

6)What factors are associated with mental health interfering with work?

```python
# Filter rows with valid work_interfere responses
df_filtered = df[df['work_interfere'].isin(['Often', 'Sometimes', 'Rarely', 'Never'])].copy()

# Create a flag: True if mental health interferes often or sometimes
df_filtered['interfere_flag'] = df_filtered['work_interfere'].isin(['Often', 'Sometimes'])

# Calculate proportion reporting interference by self_employed status
interfere_by_self_employed = df_filtered.groupby('self_employed')['interfere_flag'].mean()

# Plot
interfere_by_self_employed.plot(kind='bar', color=['skyblue', 'salmon'], figsize=(5,4))

plt.title('Mental Health Interfering with Work by Self-Employed Status')
plt.ylabel('Proportion Reporting Interference')
plt.xlabel('Self-Employed')
plt.ylim(0,1)
plt.xticks(rotation=0)
plt.tight_layout()
plt.show()
```

7)Are employees in smaller companies less aware of their mental health care options?

```python
# Filter out Unknown groups
df_filtered = df[df['company_size_group'] != 'Unknown']

# Count proportions of each response type by group
response_dist = df_filtered.groupby(['company_size_group', 'care_options']).size().unstack(fill_value=0)

# Convert to proportions
response_props = response_dist.div(response_dist.sum(axis=1), axis=0)

# Reorder rows
response_props = response_props.reindex(['Small', 'Medium', 'Large'])

# Plot
response_props[['Yes', 'No', 'Not sure']].plot(
    kind='bar', stacked=True,
    color=['seagreen', 'coral', 'gray'],
    figsize=(7, 4)
)

plt.title('Mental Health Care Options Awareness by Company Size')
plt.ylabel('Proportion of Responses')
plt.xlabel('Company Size')
plt.xticks(rotation=0)
plt.legend(title='Care Options')
plt.tight_layout()
plt.show()
```

```python
# Filter for only small companies
df_small = df[df['company_size_group'] == 'Small']

# Group by care_options
response_counts = df_small['care_options'].value_counts(normalize=True).reindex(['Yes', 'No', 'Not sure']).fillna(0)

# Plot
response_counts.plot(
    kind='bar',
    color=['seagreen', 'coral', 'gray'],
    figsize=(6, 4)
)

plt.title('Mental Health Care Options Awareness in Small Companies')
plt.ylabel('Proportion of Responses')
plt.xlabel('Response')
plt.ylim(0, 1)
plt.xticks(rotation=0)
plt.tight_layout()
plt.show()
```

8)Is there a disconnect between how mental vs. physical health are treated at work?

```python
# Filter valid responses
df_filtered = df[df['mental_vs_physical'].isin(['Yes', 'No', "Don't know"])]

# Calculate proportions of each response
response_counts = df_filtered['mental_vs_physical'].value_counts(normalize=True).reindex(['Yes', 'No', "Don't know"])

# Plot
response_counts.plot(kind='bar', color=['seagreen', 'salmon', 'gray'], figsize=(6,4))

plt.title('Perception of Mental vs Physical Health Treatment at Work')
plt.ylabel('Proportion of Employees')
plt.xlabel('Does Employer Take Mental Health as Seriously as Physical Health?')
plt.ylim(0,1)
plt.xticks(rotation=0)
plt.tight_layout()
plt.show()
```
